Q2. Draw the Silk Route in the world map. Collect all the necessary data (cities through which silk route moved and corresponding geo location) from google. Some reference available in the following site:

https://www.arcgis.com/home/item.html?id=848554492f0444a88287505f696b3212&view=list&sortOrder=desc&sortField=defaultFSOrder#data

```python
df = pd.read_csv("https://raw.githubusercontent.com/sankhla2/Data-Visualaization-Labs/main/B21AI028_DV_LAB_3/Silk_Route.csv")
df
```

```python
df['Type'].value_counts()
```

```python
df['Connected'].value_counts()
```

```python
fig = px.scatter_mapbox(df,
                        lat = 'Latitude',
                        lon = 'Longitude',
                        zoom = 3,
                        size_max = 50,
                        mapbox_style = 'open-street-map',
)
fig.update_layout(margin = dict(l =0, t=0, r =0, b=0))
fig.show()
```

```python
silk_route = pd.DataFrame(columns=['Source', 'Destination', 'Source_latitude', 'Source_longitude', 'target_latitude', 'target_longitude'])

for idx, row in df.iterrows():
    source_coordinates = (row['Latitude'], row['Longitude'])
    connected_cities = str(row['Connected']).split(', ')
    for city in connected_cities:
        destination_rows = df[df['Stop'] == city]
        if not destination_rows.empty:
            destination_row = destination_rows.iloc[0]
            destination_coords = (destination_row['Latitude'], destination_row['Longitude'])
            silk_route = pd.concat([silk_route, pd.DataFrame({
                'Source': [row['Stop']],
                'Destination': [city],
                'Source_latitude': [source_coordinates[0]],
                'Source_longitude': [source_coordinates[1]],
                'target_latitude': [destination_coords[0]],
                'target_longitude': [destination_coords[1]]
            })])

fig = px.scatter_mapbox(df,
                        lat='Latitude',
                        lon='Longitude',
                        zoom=3,
                        mapbox_style='open-street-map')

for index, row in silk_route.iterrows():
    fig.add_trace(px.line_mapbox(silk_route, lat=[row['Source_latitude'], row['target_latitude']],
                                  lon=[row['Source_longitude'], row['target_longitude']]).data[0])
fig.update_layout(margin=dict(l=0, t=0, r=0, b=0))

fig.show()
```
